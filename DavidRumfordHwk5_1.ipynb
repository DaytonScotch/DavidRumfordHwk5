```python
#importing libraries and data needed
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load iris dataset
iris = datasets.load_iris()

# Create a dataframe from iris dataset
iris_df=pd.DataFrame(iris.data)
iris_df['class']=iris.target
x = iris_df.iloc[:, [1, 2, 3, 4]].values
```

```python
#Find the optimum number of clusters for k-means classification
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300,
                    n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#Plot data on a line graph showing the "elbow method"
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```
